```
from faker import Faker
import uuid, time, json, datetime, boto3

faker = Faker()

while True:
  ts = datetime.datetime.now().strftime("%Y-%m-%dT%H:%M:%S.%f")
  data = {
    "id": str(uuid.uuid4()),
    "name": faker.name(),
    "address": faker.address(),
    "email": faker.email(),
    "birthdate": faker.date_of_birth().strftime("%Y-%m-%d"),
    "created": ts
  }
  boto3.resource('s3').Object("my-beauty-bucket", f"my-beauty-bucket/some/prefix/new-data/{ts}.json").put(
      Body=(bytes(json.dumps(data).encode('UTF-8')))
  )
#   time.sleep(1)
```
